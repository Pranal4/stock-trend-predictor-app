```python
# --- FINAL, ROBUST DATA LOADING ---
import pandas as pd
import os

# Define the file path
TICKER = "RELIANCE.NS"
DATA_PATH = "data"
RAW_DATA_FILE = os.path.join(DATA_PATH, f"{TICKER}_raw_data.csv")

# Use a more powerful read_csv command to handle the messy file
df = pd.read_csv(
    RAW_DATA_FILE,
    skiprows=3,  # Skip the top 3 lines of junk
    header=None, # Tell pandas there's no header row to read now
    names=['Date', 'Open', 'High', 'Low', 'Close', 'Volume'], # Provide the correct names ourselves
    parse_dates=['Date'], # Now, parse the 'Date' column we just named
    index_col='Date'      # And use it as the index
)

print("Data loaded successfully! The bug is fixed.")
# Display the head of the now-correctly-loaded DataFrame to prove it
display(df.head())
```

```python

import matplotlib.pyplot as plt

df['MA50'] = df['Close'].rolling(window=50).mean()
df['MA200'] = df['Close'].rolling(window=200).mean()

plt.figure(figsize=(15, 7))
df[['Close', 'MA50', 'MA200']].plot(ax=plt.gca())
plt.title('Reliance with 50-Day & 200-Day Moving Averages', fontsize=16)
plt.ylabel('Price (INR)', fontsize=12)
plt.xlabel('Date', fontsize=12)
plt.legend(['Close Price', '50-Day MA (Fast)', '200-Day MA (Slow)'])
plt.grid(True)
plt.show()
```

```python
# 1. Moving Average Spread: The difference between the fast and slow MAs.
# This numerically captures the "crossover" signals we saw in the chart.
df['MA_Spread'] = df['MA50'] - df['MA200']

# 2. Volatility: We'll use the standard deviation of the closing price over the last 20 days.
# This measures how "shaky" or "calm" the stock has been recently.
df['Volatility_20D'] = df['Close'].rolling(window=20).std()

# 3. Price Rate of Change (ROC): The percentage change in price over the last 5 days.
# This measures recent momentum.
df['ROC_5D'] = (df['Close'] - df['Close'].shift(5)) / df['Close'].shift(5) * 100

# CRITICAL STEP: These calculations create empty (NaN) values for the first 200 rows.
# Machine learning models cannot handle empty values, so we must remove these rows.
df.dropna(inplace=True)

print("Features created successfully. Let's inspect our new columns:")
display(df.head())
```

```python
# --- NEXT STEP: Target Variable Creation ---
# (Add this to the next empty cell and run it)

# Our target: 1 if the price will go UP in the next 5 days, 0 otherwise.

# Use .shift(-5) to peek 5 days into the future
df['Future_Price_5D'] = df['Close'].shift(-5)

# Create the target column by comparing future vs. current price
df['Target'] = (df['Future_Price_5D'] > df['Close']).astype(int)

# The last 5 rows have no future, creating NaNs. We must drop them.
df.dropna(inplace=True)

# We don't need the actual future price for training, just the Target signal.
df.drop(columns=['Future_Price_5D'], inplace=True)

print("Target variable 'Target' created successfully.")
print("\nHow many UP vs. DOWN days do we have?")
# normalize=True gives us the percentage, which is useful
print(df['Target'].value_counts(normalize=True))
```

```python
# --- FINAL STEP OF PHASE 2: Save the Processed Data ---
# (Add this to the final empty cell and run it)

processed_data_path = os.path.join(DATA_PATH, f'{TICKER}_processed_data.csv')
df.to_csv(processed_data_path)

print(f"PHASE 2 COMPLETE!")
print(f"Processed data saved successfully to: {processed_data_path}")
```
